### TEPT: Primeiro Dataset

```python
tept = pd.read_csv('tept_df.csv')
tept.head(5)

# Armazena em uma variável o dataset construído anteriormente e exibe as 5 primeiras observações.
```

```python
tept2 = tept.replace({0: 'não', 1: 'sim'})
tept2.head(5)

# Substitui, em todas as colunas, os valores 0 e 1 por não e sim, respectivamente;
# Exibe as 5 primeiras observações.
```

```python
tept3 = tept2.rename(columns={'Unnamed: 0': 'id'})
tept3.head(5)

# Altera nome da coluna Unnamed: 0 para id e exibe as 5 primeiras observações.
```

```python
tept4 = tept3.drop(['Unnamed: 0.1', 'tept'], axis=1)
tept4.head(5)

# Exclui as colunas Unnamed: 0.1 e tept do dataframe e exibe as 5 primeiras observações.
```

```python
tept4.to_csv('tept.csv', index=False)

# Salva o dataframe como um arquivo csv, sem o índice.
```

### TEPT: Segundo Dataset

```python
tept_grupo = pd.read_csv('tept_df.csv')
tept_grupo.head(5)

# Armazena em uma variável o dataset construído anteriormente e exibe as 5 primeiras observações.
```

```python
tept_grupo['crit_a'] = tept_grupo[['a1', 'a2', 'a3', 'a4']].sum(axis=1)
tept_grupo['crit_b'] = tept_grupo[['b1', 'b2', 'b3', 'b4', 'b5']].sum(axis=1)
tept_grupo['crit_c'] = tept_grupo[['c1', 'c2']].sum(axis=1)
tept_grupo['crit_d'] = tept_grupo[['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7']].sum(axis=1)
tept_grupo['crit_e'] = tept_grupo[['e1', 'e2', 'e3', 'e4', 'e5', 'e6']].sum(axis=1)

# Cria novas colunas com as somas dos critérios diagnósticos presentes, de acordo com suas categorias.
```

```python
novo_df = tept_grupo[['crit_a', 'crit_b', 'crit_c', 'crit_d', 'crit_e', 'tept']]
novo_df

# Cria e exibe um novo dataframe a partir da seleção das novas colunas + tept.
```
