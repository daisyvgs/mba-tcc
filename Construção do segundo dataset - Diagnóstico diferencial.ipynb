### Transtorno de Estresse Pós-Traumático

```python
tept_variacao = np.arange(1, 501)
tept_humor = ['medo', 'desespero', 'horror', 'raiva', 'culpa', 'vergonha', 'anedonia', 'disforia', 'sofrimento prolongado', 'sofrimento intenso', 'irritacao', 'emocoes positivas ausentes']
tept_cognicao = ['lembranca recorrente', 'revivencia', 'pesadelos', 'amnesia dissociativa', 'expectativa negativa', 'cognicao erronea', 'dificuldade de concentracao', 'preocupacao', 'ideias paranoides', 'luto complicado']
tept_comportamento = ['reativo', 'evitativo', 'agressivo', 'irritadico', 'imprudente', 'autodestrutivo', 'automutilante', 'suicida']
tept_percepcao = ['intrusao visual', 'intrusao sensorial', 'pseudoalucinacoes auditivas', 'flashbacks', 'despersonalizacao', 'desrealizacao']
tept_fisiologico = ['reatividade fisiologica', 'resposta de sobressalto exagerada', 'perturbacao do sono']
tept_risco = ['evento traumatico', 'transtorno mental anterior', 'estrategias de enfrentamento inapropriadas']
tept_duracao = ['> 2 meses']
tept_resultado = ['transtorno de estresse pos-traumatico']

# Armazena as características diagnósticas em arrays, sendo que cada array corresponde a uma variável do dataset.
```

```python
max_length = max(len(tept_variacao), len(tept_humor), len(tept_cognicao), len(tept_comportamento), len(tept_percepcao), len(tept_fisiologico), len(tept_risco), len(tept_duracao), len(tept_resultado))

# Armazena o comprimento máximo de cada array.
```

```python
tept_variacao_extended = list(islice(cycle(tept_variacao), max_length))
tept_humor_extended = list(islice(cycle(tept_humor), max_length))
tept_cognicao_extended = list(islice(cycle(tept_cognicao), max_length))
tept_comportamento_extended = list(islice(cycle(tept_comportamento), max_length))
tept_percepcao_extended = list(islice(cycle(tept_percepcao), max_length))
tept_fisiologico_extended = list(islice(cycle(tept_fisiologico), max_length))
tept_risco_extended = list(islice(cycle(tept_risco), max_length))
tept_duracao_extended = list(islice(cycle(tept_duracao), max_length))
tept_resultado_extended = list(islice(cycle(tept_resultado), max_length))

# Armazena a repetição dos arrays menores, considerando a maior variável (tept_variacao);
# Possibilita a variação das observações e garante que todas as colunas do dataset tenham o mesmo tamanho.
```

```python
tept_df = pd.DataFrame({
    'VARIACAO': tept_variacao_extended,
    'HUMOR': tept_humor_extended,
    'COGNICAO': tept_cognicao_extended,
    'COMPORTAMENTO': tept_comportamento_extended,
    'PERCEPCAO': tept_percepcao_extended,
    'FISIOLOGICO': tept_fisiologico_extended,
    'RISCO': tept_risco_extended,
    'DURACAO': tept_duracao_extended,
    'RESULTADO': tept_resultado_extended
})

# Cria um dataframe no qual cada array corresponde a uma coluna.
```

```python
tept_df.tail(5)
```
